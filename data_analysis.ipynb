### Step 2. Prepare Phase:
Sršen suggested that I utilize openly accessible data that analyzes the everyday schedules of individuals who use smart devices. She directed me toward the subsequent dataset:

The Mobius-hosted data set available on Kaggle under the CC0 Public Domain license consists of personal fitness tracker data from thirty Fitbit users who voluntarily disclosed their minute-by-minute heart rate, sleep, and physical activity monitoring information. The data encompasses details on daily activity, steps taken, and heart rate, enabling thorough analysis of users' routines.There are 18 CSV files available on Kaggle that contain data from the FitBit fitness tracker which can be accessed for free.

During the Prepare step, we identify the data being utilized and its limitations.

- The data was collected in 2016, which was seven years ago. As time has passed, users may have altered their daily routines, including their eating, exercise, and sleep habits. Therefore, the data may not be current or relevant.

- It is not appropriate to assume that the entire fitness market can be accurately represented based on a sample size of only 30 FitBit users.

- We cannot guarantee the accuracy or reliability of the data since it was collected through a survey.

### Is Data ROCCC?
ROCCC, an acronym for Reliable, Original, Comprehensive, Current, and Cited, serves as a dependable and trustworthy source of information.

- Reliable — LOW — Having only 30 responders makes it unreliable.

- Original — LOW — The service being referred to is provided by a third party, specifically Amazon Mechanical Turk.

- Comprehensive — MED — The device's parameters align with those typically found in Bellabeat devices.

- Current — LOW — Data that is 7 years old may no longer be applicable or significant.

- Cited — LOW — The data was obtained from an unspecified third party. Overall, the dataset is considered to have poor-quality data, and it is not recommended to rely on it for making business recommendations.

### Step 5. Share Phase:
In this step, I'll produce some visual aids based on our analysis and project goals.

```python
# Set the order of the days of the week
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
daily_activity["Day_of_week"] = pd.Categorical(daily_activity["Day_of_week"], categories=day_order, ordered=True)

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x="Day_of_week", data=daily_activity, color='#0077b6')

# Set the labels and title
plt.xlabel("Day of Week")
plt.ylabel("Count")
plt.title("No. of times users used tracker across week")

# Show the plot
plt.show()
```

FitBit app usage is highest on Sundays, Mondays, and Tuesdays, possibly due to people being busier on weekends and having less time to monitor physical activity.

```python
da_days = daily_activity.groupby('Day_of_week').agg({'TotalDistance': 'sum'}).reset_index().sort_values('TotalDistance', ascending=False)

fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x='Day_of_week', y='TotalDistance', data=da_days, color='#023e8a', ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Total Distance During Each Day')
ax.set_ylabel('Total Distance Sum')
ax.set_xlabel('Days')

plt.show()
```

In contrast, we can observe that Tuesday, Wednesday, and Thursday have the highest total distance sum among all days. It is worth noting that Tuesday appears frequently in both charts, while Sunday and Monday have the lowest values.

```python
mean_steps = daily_activity['TotalSteps'].mean()
mean_calories = daily_activity['Calories'].mean()

plt.figure(figsize=(12, 8))
palette = 'magma'
cmap = sns.color_palette(palette, as_cmap=True).reversed()

sns.scatterplot(data=daily_activity, x='TotalSteps', y='Calories', hue='Calories', palette=cmap)
plt.axhline(mean_calories, color='red', linewidth=1.0)
plt.axvline(mean_steps, color='blue', linewidth=1.0)
plt.text(10000, 500, 'Average Steps', rotation=90)
plt.text(29000, 2500, 'Average Calories')
plt.xlabel('Steps Taken')
plt.ylabel('Calories Burned')
plt.title('Calories burned for every step taken')

plt.show()
```

- The scatter plot indicates a positive correlation, despite a few outliers that deviate from the trend.

- The data clearly shows a positive correlation between the number of steps taken and the amount of calories burned.

```python
plt.figure(figsize=(12, 8))
palette = 'magma'
cmap = sns.color_palette(palette, as_cmap=True).reversed()

sns.scatterplot(data=daily_activity, x='TotalSteps', y='SedentaryMinutes', hue='Calories', palette=cmap)
sns.regplot(data=daily_activity, x='TotalSteps', y='SedentaryMinutes', scatter=False, color='green', lowess=True)
plt.xlabel('Total Steps')
plt.ylabel('Sedentary Minutes')
plt.title('Total Steps vs Sedentary Minutes')

plt.show()
```

- As the step count surpasses 10,000, the relationship between them remains fairly constant, despite being inverse for fewer steps.

- I noticed that after taking 15000 steps, there was a slight shift towards a positive relationship between steps and sedentary minutes.
